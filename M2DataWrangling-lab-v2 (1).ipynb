### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
# Identify columns with the highest number of missing values from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

import pandas as pd

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(url)

missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")

# Calculate the percentage of missing values for each column
missing_percentage = missing_data.sum() / len(df) * 100

# Sort the columns by the percentage of missing values in descending order
sorted_missing_percentage = missing_percentage.sort_values(ascending=False)

# Get the columns with the highest number of missing values
# For example, to get the top 5 columns with the most missing values:
top_missing_columns = sorted_missing_percentage.head(5)

print("Columns with the highest number of missing values:")
top_missing_columns
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
# Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv')

# Drop rows with NaN values in the target column (optional, but recommended before scaling)
df.dropna(subset=['ConvertedCompYearly'], inplace=True)

# Select the column to scale
data_to_scale = df[['ConvertedCompYearly']]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Apply Min-Max Scaling
df['ConvertedCompYearly_scaled'] = scaler.fit_transform(data_to_scale)

# Display the first few rows with the new scaled column
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_scaled']].head())
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
from IPython import get_ipython
from IPython.display import display
# %%
# prompt: Create a new column `ExperienceLevel` based on the `YearsCodePro` column from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

!wget https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

import pandas as pd

df = pd.read_csv('survey-data.csv')

# Convert 'YearsCodePro' to numeric, coercing errors to NaN
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Define a function to categorize experience level
def categorize_experience(years):
    if pd.isna(years):
        return 'Unknown'
    elif years < 5:
        return 'Entry-level'
    elif 5 <= years < 15:
        return 'Mid-level'
    else:
        return 'Senior-level'

# Apply the function to create the new column
df['ExperienceLevel'] = df['YearsCodePro'].apply(categorize_experience)

# Display the first few rows with the new column
print(df[['YearsCodePro', 'ExperienceLevel']].head(20))
```
