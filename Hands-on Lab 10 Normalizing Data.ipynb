##### Task 5: Normalize ConvertedCompYearly using Min-Max Scaling.
Min-Max Scaling brings all values in a column to a 0-1 range, making it useful for comparing data across different scales. Here, you will apply Min-Max normalization to the ConvertedCompYearly column, creating a new column ConvertedCompYearly_MinMax with normalized values.


```python
## Write your code here
# prompt: Normalize ConvertedCompYearly using Min-Max Scaling from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv
!pip install scikit-learn
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
df = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv')

# Drop rows with NaN in ConvertedCompYearly before scaling
df_cleaned = df.dropna(subset=['ConvertedCompYearly'])

# Select the column to normalize
data_to_normalize = df_cleaned[['ConvertedCompYearly']]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
df_cleaned['ConvertedCompYearly_Normalized'] = scaler.fit_transform(data_to_normalize)

# Display the first few rows of the dataframe with the new column
print(df_cleaned[['ConvertedCompYearly', 'ConvertedCompYearly_Normalized']].head())
```

##### Task 6: Apply Z-score Normalization to `ConvertedCompYearly`.

Z-score normalization standardizes values by converting them to a distribution with a mean of 0 and a standard deviation of 1. This method is helpful for datasets with a Gaussian (normal) distribution. Here, you’ll calculate Z-scores for the ConvertedCompYearly column, saving the results in a new column ConvertedCompYearly_Zscore.


```python
## Write your code here
# Apply Z-score normalization
df['ConvertedCompYearly_Zscore'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].mean()) / df['ConvertedCompYearly'].std()

# Display the first few rows with the new column
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Zscore']].head())
```

### Section 4: Visualization of Normalized Data
##### Task 7: Visualize the distribution of `ConvertedCompYearly`, `ConvertedCompYearly_Normalized`, and `ConvertedCompYearly_Zscore`

Visualization helps you understand how normalization changes the data distribution. In this task, create histograms for the original ConvertedCompYearly, as well as its normalized versions (ConvertedCompYearly_MinMax and ConvertedCompYearly_Zscore). This will help you compare how each normalization technique affects the data range and distribution.


```python
## Write your code here
# Visualize the distribution of `ConvertedCompYearly`, `ConvertedCompYearly_Normalized`, and `ConvertedCompYearly_Zscore` Visualization helps you understand how normalization changes the data distribution. In this task, create histograms for the original ConvertedCompYearly, as well as its normalized versions (ConvertedCompYearly_MinMax and ConvertedCompYearly_Zscore). This will help you compare how each normalization technique affects the data range and distribution.
# from https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv

import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv")

# Select the relevant column
df_comp = df[['ConvertedCompYearly']].copy()

# Drop missing values
df_comp.dropna(inplace=True)

# Min-Max Normalization
scaler_minmax = MinMaxScaler()
df_comp['ConvertedCompYearly_MinMax'] = scaler_minmax.fit_transform(df_comp[['ConvertedCompYearly']])

# Z-score Standardization
scaler_standard = StandardScaler()
df_comp['ConvertedCompYearly_Zscore'] = scaler_standard.fit_transform(df_comp[['ConvertedCompYearly']])

# Create histograms
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df_comp['ConvertedCompYearly'], bins=30, edgecolor='black')
plt.title('Original ConvertedCompYearly')
plt.xlabel('ConvertedCompYearly')
plt.ylabel('Frequency')

plt.subplot(1, 3, 2)
plt.hist(df_comp['ConvertedCompYearly_MinMax'], bins=30, edgecolor='black')
plt.title('Min-Max Normalized ConvertedCompYearly')
plt.xlabel('ConvertedCompYearly_Normalized')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(df_comp['ConvertedCompYearly_Zscore'], bins=30, edgecolor='black')
plt.title('Z-score Standardized ConvertedCompYearly')
plt.xlabel('ConvertedCompYearly_Zscore')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
